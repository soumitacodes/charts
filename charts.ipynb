```python
import matplotlib.pyplot as plt
```

line chat

```python
x_line=[1,2,3,4,5]
y_line=[10,12,8,14,20]
#plotting the linen chart
plt.figure(figsize=(6,4))
plt.plot(x_line,y_line,color='blue',marker="o",label='line data')
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.legend()
plt.grid(True)
plt.show()
```

BAR CHART

```python
catagories=['Apple','banana','cherry','orange']
values=[30,50,87,20]
plt.figure(figsize=(6,4))
plt.bar(catagories,values,color='orange')
plt.title('bar chart')
plt.xlabel('fruits')
plt.ylabel('Quantity')
plt.show()
```

SCATTER plot

```python
x_scatter=[5,7,8,7,2,17,2,9]
y_scatter=[99,86,87,88,100,86,103,87]
plt.figure(figsize=(6,4))
plt.scatter(x_scatter,y_scatter,color='red',marker='x')
plt.title('scatter chart')
plt.xlabel('age')
plt.ylabel('score')
plt.show()
```

pie chart

```python
catagories=['Apple','banana','cherry','orange']
values=[30,50,87,20]
plt.figure(figsize=(6,4))
plt.pie(values,labels=catagories,autopct='%1.1f%%',startangle=90)
plt.title('pie chart')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
```

histrogram

```python
data=sns.load_dataset('tips')
plt.figure(figsize=(6,4))
sns.histplot(data['total_bill'],bins=10,color='skyblue',edgecolor='black')
plt.title('Histogram og total bills');
plt.xlabel('bill')
plt.ylabel('frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='day',y='total_bill',data=data,palette='Pastel1')
plt.title('bill')
plt.xlabel('day of')
plt.ylabel('total bill')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='day',y='total_bill',data=data,palette='Set2')
plt.title('bill')
plt.xlabel('day of')
plt.ylabel('total bill')
plt.tight_layout()
plt.show()
```
